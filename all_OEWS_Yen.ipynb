```python
import pandas as pd
```

```python
all_oews_df = pd.read_excel("Data/all_data_M_2023.xlsx")
all_oews_df.head()
```

```python
# Area type: 1= U.S.; 2= State; 3= U.S. Territory; 4= Metropolitan Statistical Area (MSA) or New England City and Town Area (NECTA); 6= Nonmetropolitan Area 
# Narrow down the data to State level
state_oews_df = all_oews_df.loc[all_oews_df['AREA_TYPE'] == 2, :]
state_oews_df
```

```python
# Check the data data type and non-null count (preparation for transformation)
state_oews_df.info()
```

```python
# Get a list of columns
state_oews_df.columns
```

```python
# Get rid of unneccssary columns for state level data
# Columns "PCT_TOTAl" and "PCT_RPT" are only available for the national industry estimates
# Columns "ANNUAL" and "HOURLY" contains "TRUE" if only annual/hourly wages are released. The OEWS program releases only annual wages for some occupations that typically work fewer than 2,080 hours per year
# del df["column you want to delete"]
# df = df.drop(columns=['A', 'C'])
# selected_columns = df[['A', 'C']]

clean_oews_df = state_oews_df[['AREA', 'AREA_TITLE', 'PRIM_STATE','OCC_CODE', 'OCC_TITLE', 'O_GROUP',\
     'TOT_EMP', 'EMP_PRSE', 'JOBS_1000', 'LOC_QUOTIENT','H_MEAN', 'A_MEAN', 'MEAN_PRSE', 'H_PCT25', 'H_MEDIAN', 'H_PCT75','A_PCT25', 'A_MEDIAN', 'A_PCT75']]

clean_oews_df.head()
```

```python
# Rename the columns for clear understanding
# df.rename(columns={"old name":"new name", "old name":"new name"})

renamed_oews_df = clean_oews_df.rename(columns={"AREA": "State_FIPS_Code", "AREA_TITLE": "State", "OCC_CODE": "OCC_Code", "OCC_TITLE": "OCC_Title", "O_GROUP": "OCC_Level", "TOT_EMP": "Total_Employment",\
    "EMP_PRSE": "Employment_PRSE", "JOBS_1000": "Jobs_1000", "LOC_QUOTIENT": "Location_Quotient", "H_MEAN": "Hourly_Wage_Mean","A_MEAN":"Annual_Wage_Mean", "MEAN_PRSE": "Mean_PRSE",\
         "H_PCT25": "Hourly_PCT25", "H_MEDIAN": "Hourly_PCT50", "H_PCT75": "Hourly_PCT75", "A_PCT25": "Annual_PCT25", "A_MEDIAN": "Annual_PCT50", "A_PCT75": "Annual_PCT75"
    })

renamed_oews_df.head(100)
```

```python
group_total = clean_oews_df.loc[clean_oews_df['O_GROUP'] == 'total', :]
group_total_count = len(group_total['OCC_TITLE'].unique())

group_major = clean_oews_df.loc[clean_oews_df['O_GROUP'] == 'major', :]
group_major_count = len(group_major['OCC_TITLE'].unique())

group_detailed = clean_oews_df.loc[clean_oews_df['O_GROUP'] == 'detailed', :]
group_detailed_count = len(group_detailed['OCC_TITLE'].unique())

print(f'There are {group_total_count} unique title values under group total.')
print(f'There are {group_major_count} unique title values under group major.')
print(f'There are {group_detailed_count} unique title values under group detailed.')
```
