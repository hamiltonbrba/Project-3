```python
import pandas as pd

# Load the Excel file (adjust the path as needed)
file_path = 'Data/Data_Scientist_Analyst.xlsx'

# Read the Excel file into a DataFrame
df = pd.read_excel(file_path)

# Display the first few rows to understand the structure of the data
print(df.head())
```

```python
# Load the Excel file again, skipping the first few rows to get to the actual data
df_cleaned = pd.read_excel(file_path, skiprows=4)

# Display the first few rows of the cleaned DataFrame
df_cleaned.head()
```

```python
# Set the first row as the header
df_cleaned.columns = df_cleaned.iloc[0]
df_cleaned = df_cleaned.drop(0)

# Display the cleaned DataFrame
df_cleaned.head()
```

```python
# Renaming the columns to be more specific and appropriate
df_cleaned.columns = [
    'State', 
    'Employment', 
    'Hourly Mean Wage', 
    'Annual Mean Wage', 
    'Hourly Median Wage'
]

# Display the updated DataFrame with the new column names
df_cleaned.head()
```

```python
# Remove numbers and parentheses from the 'State' column to clean up the state names
df_cleaned['State'] = df_cleaned['State'].str.replace(r'\(.*\)', '', regex=True).str.strip()

# Display the updated DataFrame
df_cleaned.head()
```
